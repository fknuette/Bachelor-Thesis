```python
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd


# Get the data
data = pd.read_csv('/pfs/work7/workspace/scratch/ma_fknuette-project_GRANDE/data/result/cifar10/dataGeneral.csv')

data_short = data[['clean', 'fgsm', 'pgd', 'apgd', 'square']]

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Prepare data for scatter plots
x1, y1 = data_short['clean'], data_short['fgsm']
x2, y2 = data_short['clean'], data_short['pgd']
x3, y3 = data_short['clean'], data_short['apgd']
x4, y4 = data_short['clean'], data_short['square']

# Update scatter plot data
axs[0, 0].scatter(x1, y1, color='blue')
axs[0, 0].set_title(f'Clean vs FGSM : {data_short["clean"].corr(data_short["fgsm"]):.4f}')

axs[0, 1].scatter(x2, y2, color='green')
axs[0, 1].set_title(f'Clean vs PGD : {data_short["clean"].corr(data_short["pgd"]):.4f}')

axs[1, 0].scatter(x3, y3, color='red')
axs[1, 0].set_title(f'Clean vs APGD : {data_short["clean"].corr(data_short["apgd"]):.4f}')

axs[1, 1].scatter(x4, y4, color='purple')
axs[1, 1].set_title(f'Clean vs Square : {data_short["clean"].corr(data_short["square"]):.4f}')


# Add labels to axes
for ax in axs.flat:
    ax.set(xlabel='Clean', ylabel='Adversarial')

# Adjust layout
plt.tight_layout()
plt.show()
```
